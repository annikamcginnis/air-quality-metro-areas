```python
import pandas as pd

all_data = []  # list to collect each year's data

for year in range(1980, 2025):
    filename = f'daily_88101_{year}.csv'
    try:
        df = pd.read_csv(filename)
        all_data.append(df)
    except FileNotFoundError:
        print(f"File not found: {filename}")

# Concatenate all years into a single DataFrame
all_years_df = pd.concat(all_data, ignore_index=True)
```

## Fuzzy matching the CBSA names from both datasets

```python
# Get unique CBSA names from the air quality data
cbsa_names = all_years_df['CBSA Name'].dropna().unique()

# Get metro names from your reference DataFrame
metro_names = metros['matched_metro_area_2023'].dropna().unique()
```

```python
from fuzzywuzzy import process
```

```python
# Create a mapping: each CBSA Name → best match in metros
fuzzy_matches = []
for name in cbsa_names:
    match, score = process.extractOne(name, metro_names)
    fuzzy_matches.append({'cbsa_name': name, 'matched_metro': match, 'score': score})

matches_df = pd.DataFrame(fuzzy_matches)
```

```python
#pd.set_option('display.max_rows', None)
```

```python
matches_df = matches_df[matches_df['score'] > 92].sort_values(by='score')
```

```python
matches_df
```

```python
matches_df['CBSA Name'] = matches_df['cbsa_name']
```

```python
metros['CBSA Name'] = metros['matched_metro_area_2023']
```

```python
matches_df_merged = matches_df.merge(metros, on='CBSA Name', how='left')
```

```python
matches_df_merged
```

## Filtering the air quality dataset only to metropolitan areas listed in the metros data

```python
# First, extract the list of matched metro names from your matches_df
matched_metros = matches_df['cbsa_name'].unique()

# Now filter all_years_df using that list
filtered_df = all_years_df[all_years_df['CBSA Name'].isin(matched_metros)]
```

```python
filtered_df.head()
```

```python
filtered_df.shape
```

## Filtering by 24-HOUR readings and averaging the Air Quality Index for all local sites within the same metro area per year

```python
readings_24_hr = filtered_df[filtered_df['Sample Duration'] == '24 HOUR']
```

```python
readings_24_hr['CBSA Name'].value_counts().shape
```

```python
site_year_avg = readings_24_hr.groupby(['Local Site Name', 'Year'], as_index=False)['AQI'].mean()
```

```python
site_year_avg
```

```python
site_to_cbsa = readings_24_hr[['Local Site Name', 'CBSA Name']].drop_duplicates()
```

```python
site_year_avg = site_year_avg.merge(site_to_cbsa, on='Local Site Name', how='left')
```

```python
site_year_avg
```

```python
cbsa_year_avg = site_year_avg.groupby(['CBSA Name', 'Year'], as_index=False)['AQI'].mean()
```

```python
cbsa_year_avg.head(30)
```

```python
pivot_df = cbsa_year_avg.pivot(index='CBSA Name', columns='Year', values='AQI')
```

```python
pivot_df = pivot_df.reset_index()
```

```python
pivot_df
```

```python
metros
```

## Merging in the population percent change column

```python
pop_pct_change = metros[['CBSA Name','pop_pct_change']]
```

```python
pop_pct_change
```

```python
merge = pivot_df.merge(pop_pct_change, on='CBSA Name', how='left')
```

```python
merge
```

```python
# Manually fill in missing pop_pct_change values
merge.loc[merge['CBSA Name'] == 'Chicago-Naperville-Elgin, IL-IN-WI', 'pop_pct_change'] = -0.016487
merge.loc[merge['CBSA Name'] == 'Grand Rapids-Wyoming, MI', 'pop_pct_change'] = 0.145922
merge.loc[merge['CBSA Name'] == 'New York-Newark-Jersey City, NY-NJ-PA', 'pop_pct_change'] = -0.005453
merge.loc[merge['CBSA Name'] == 'Raleigh, NC', 'pop_pct_change'] = 0.218577
merge.loc[merge['CBSA Name'] == 'Salt Lake City, UT', 'pop_pct_change'] = 0.122542
```

```python
merge
```

```python
merge = merge.sort_values(by='pop_pct_change',ascending=False)
```

```python
merge
```

```python
merge = merge.drop(columns='pop_pct_change')
```

## Turning the data into long format for import to Observable as json

```python
# Reshape with melt (keeping 'CBSA Name' and 'pop_pct_change' intact)
tidy_df = merge.melt(
    id_vars="CBSA Name",
    var_name="year",
    value_name="value"
)
```

```python
tidy_df
```

```python
tidy_df.dtypes
```

```python
import numpy as np

tidy_df.loc[tidy_df["year"].isin([1997, 1998]), "value"] = np.nan
```

```python
tidy_df = tidy_df.fillna(0)
```

```python
tidy_df['date'] = pd.to_datetime(tidy_df['year'].astype(str) + '-01-01')
```

```python
tidy_df.to_csv('top_growing_large_metros_pm25.csv',index=False)
```
