## Calculating the percentage change in population between 2014 and 2023

```python
df['population_2014'] = df['population_2014'].str.replace(',', '').astype(int)
df['population_2023'] = df['population_2023'].str.replace(',', '').astype(int)
```

```python
df['population_2014'] = df['population_2014'].astype(int)
```

```python
df['population_2023'] = df['population_2023'].astype(int)
```

```python
df['pop_pct_change'] = (df['population_2023'] - df['population_2014'])/df['population_2014']
```

```python
fastest_growing_large_metros = df[df['population_2023'] > 1000000].sort_values(by='pop_pct_change',ascending=False)
```
